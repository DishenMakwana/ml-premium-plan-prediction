```python
import pandas as pd
```

```python
df_sales = pd.read_excel('./linechart.xlsx')
df_sales.head()
```

```python
from matplotlib import pyplot as plt

plt.plot(df_sales['Quarter'], df_sales['Fridge'], label='Fridge')
plt.plot(df_sales['Quarter'], df_sales['Washing Machine'], label='Washing Machine')
plt.plot(df_sales['Quarter'], df_sales['Dishwasher'], label='Dishwasher')

plt.title('Sales of Appliances')
plt.xlabel('Financial Quarter')
plt.ylabel('Revenue in million $')

plt.legend()

plt.show()
```

```python
total_sales = df_sales[['Fridge', 'Washing Machine', 'Dishwasher']].sum()
```

```python
plt.pie(total_sales, labels=total_sales.index, autopct='%1.1f%%', shadow=True, startangle=140)
plt.show()
```

```python
df_sales.plot(kind='bar', x="Quarter")
plt.xticks(rotation=45)

plt.show()
```
