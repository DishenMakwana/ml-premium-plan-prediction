```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df_s = pd.read_csv('shoe_sales.csv')
df_s.head()
```

```python
df_s.shape
```

```python
df_s.describe()
```

```python
df = df_s.copy()
```

```python
df_n = df[df['brand'] == 'Nike']
```

```python
df_n.describe()
```

```python
df_n[df_n.sold_qty.isnull()]
```

```python
df_n.sold_qty.fillna(20, inplace=True)
```

```python
df_n.loc[[32,48]]
```

```python
df_a = df[df['brand'] == 'Adidas']
```

```python
df_a.describe()
```

```python
df_a[df_a.sold_qty.isnull()]
```

```python
df_n.sold_qty.sum()
```

```python
df_a.sold_qty.sum()
```

```python
df_a.sold_qty.quantile([0.90])
```

```python
df_a[df_a.sold_qty > 17.1]
```

```python
plt.figure(figsize=(15, 6))
dates = df_n['date']
plt.plot(dates, df_n['sold_qty'], label='Nike', color='blue')
plt.plot(dates, df_a['sold_qty'], label='Adidas', color='red')
plt.xlabel('Date')
plt.ylabel('Sold Quantity')
plt.title('Sold Quantity of Nike and Adidas')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

```python
val = round(df_a.sold_qty.median())
val
```

```python
df_a.sold_qty.replace(689, val, inplace=True)
```

```python
plt.figure(figsize=(15, 6))
dates = df_n['date']
plt.plot(dates, df_n['sold_qty'], label='Nike', color='blue')
plt.plot(dates, df_a['sold_qty'], label='Adidas', color='red')
plt.xlabel('Date')
plt.ylabel('Sold Quantity')
plt.title('Sold Quantity of Nike and Adidas')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```
